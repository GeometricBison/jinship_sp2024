```python
import pandas as pd
from datasets import load_dataset

dataset = load_dataset("hugginglearners/netflix-shows", split="train")
df = pd.DataFrame(dataset)
```

```python
# Create bucket queue from the genres used in 'listed_in'
genre_lists = df['listed_in'].str.split(", ")
all_genres = set()
for genres in genre_lists:
    all_genres.update(genres)
bucket_queue = {genre: [] for genre in all_genres}
```

```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load pre-trained GPT-2 model and tokenizer
model_name = "gpt2-medium"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

# Define genre
genre = "Horror"

# Generate and print horror descriptions
for i in range(1):  # Generate 5 descriptions
    prompt = f"The genre is {genre}. Describe 5 different {genre} TV shows through a short description: "
    input_ids = tokenizer.encode(prompt, return_tensors="pt")

    # Generate description using the model
    max_length = 100  # Maximum length of the generated description
    output = model.generate(input_ids, max_length=max_length, num_return_sequences=1, pad_token_id=tokenizer.eos_token_id)
    generated_description = tokenizer.decode(output[0], skip_special_tokens=True)

    print(f"Description {i+1}: {generated_description}\n")
```

```python
from transformers import pipeline, set_seed
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load pre-trained GPT-2 model and tokenizer
model_name = "gpt2-medium"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

# Set up text generation pipeline
generator = pipeline('text-generation', model=model, tokenizer=tokenizer)

# Set random seed for reproducibility
set_seed(42)

# Custom prompt for generating horror movie descriptions
prompt = "Given the genre is horror, can you generate a description that describes a scary movie or tv show. for example: a ghost haunts a family in their new mansion :\n1. "

# Generate horror movie descriptions
generated_descriptions = generator(prompt, max_length=100, num_return_sequences=5)

# Extract and print the generated horror movie descriptions
for i, description in enumerate(generated_descriptions, start=1):
    print(f"Description {i}: {description['generated_text']}")
```

```python
################################################################################
################################################################################
################################################################################

import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load pre-trained GPT-2 model and tokenizer
model_name = "gpt2"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

# Define genre descriptions
genres = list(bucket_queue.keys())

# Generate and store descriptions for each genre
genre_descriptions = {}
for genre in genres:
    prompt = f"The genre is {genre}. Describe a typical {genre} TV show: "
    input_ids = tokenizer.encode(prompt, return_tensors="pt")

    # Generate descriptions using the model
    max_length = 50  # Maximum length of the generated description
    num_descriptions = 3  # Number of descriptions to generate for each genre
    generated_descriptions = []
    for _ in range(num_descriptions):
        output = model.generate(input_ids, max_length=max_length, num_return_sequences=1, pad_token_id=tokenizer.eos_token_id)
        generated_description = tokenizer.decode(output[0], skip_special_tokens=True)
        generated_descriptions.append(generated_description)

    genre_descriptions[genre] = generated_descriptions

# Print generated description for the genres
for genre, descriptions in genre_descriptions.items():
    print(f"Genre: {genre}")
    for i, description in enumerate(descriptions, start=1):
        print(f"Description {i}: {description}")
    print("\n")
```

```python
from fuzzywuzzy import fuzz

# Function to find the most similar genre for a given description
def find_genre(description):
    max_similarity = 0
    most_similar_genre = None

    for genre, descriptions in genre_descriptions.items():
        for gen_desc in descriptions:
            similarity = fuzz.token_sort_ratio(description.lower(), gen_desc.lower())
            if similarity > max_similarity:
                max_similarity = similarity
                most_similar_genre = genre

    return most_similar_genre

# Iterate through the dataset to label rows with corresponding buckets
df_small = df.head(500)
for index, row in df_small.iterrows():
    description = row['description']
    genre = find_genre(description)
    df_small.at[index, 'bucket'] = genre
```

```python
from collections import defaultdict

# Initialize counters for correct and total predictions for each bucket
correct_predictions = defaultdict(int)
total_predictions = defaultdict(int)

# Loop through each row in the dataset
for index, row in df_small.iterrows():
    predicted_genre = row['bucket']  # Predicted genre
    listed_in = row['listed_in'].split(", ")  # Actual genres listed in the 'listed_in' column

    # Check if the predicted genre matches any of the actual genres
    for genre in listed_in:
        total_predictions[genre] += 1  # Increment total predictions for the actual genre
        if predicted_genre == genre:
            correct_predictions[genre] += 1  # Increment correct predictions if the predicted genre matches the actual genre

# Calculate accuracy for each bucket
accuracy_per_bucket = {}
for genre, total in total_predictions.items():
    if total > 0:
        accuracy_per_bucket[genre] = correct_predictions[genre] / total

# Overall accuracy
overall_accuracy = sum(correct_predictions.values()) / sum(total_predictions.values())

print("Accuracy per Bucket:")
for genre, accuracy in accuracy_per_bucket.items():
    print(f"{genre}: {accuracy:.2f}")

print(f"\nOverall Accuracy: {overall_accuracy:.2f}")
```
